```python
class VeeamClient:
    _API_VERSION = "1.0-rev2"

    _refresh_token = namedtuple('RefreshToken', ['value', 'expire'])
    _access_token  = namedtuple('AccessToken', ['value', 'expire'])

    def __init__(self, username, password, base_url):
        self.username = username
        self.password = password
        self.BASE_URL = base_url

        if username and password:
            self._login()
    
    def _login(self):
        # Get the access token
        res = requests.post(
            url = f"{self.BASE_URL}/api/oauth2/token",
            headers= {
                "x-api-version": self._API_VERSION
            },
            data={
                "grant_type": "password",
                "username": self.username,
                "password": self.password
            }
        )

        if(res.sstatus_code == 200):
            self._refresh_token.value = res.json()["refresh_token"]
            self._access_token.value = res.json()["access_token"]
    
    def acess_code(self):
        pass
        


```
